```python
import numpy as np
import matplotlib.pyplot as plt

def differential_equation(v, x, t, m, k, mu_s, g):
    dxdt = v
    dvdt = (-k * x - mu_s * m * g * np.sign(v)) / m
    return dvdt, dxdt

def runge_kutta_method(f, v0, x0, t, m, k, mu_s, g):
    dt = t[1] - t[0]
    v = np.zeros_like(t)
    x = np.zeros_like(t)
    v[0] = v0
    x[0] = x0

    for i in range(1, len(t)):
        k1v, k1x = f(v[i-1], x[i-1], t[i-1], m, k, mu_s, g)
        k2v, k2x = f(v[i-1] + 0.5*k1v, x[i-1] + 0.5*k1x, t[i-1] + 0.5*dt, m, k, mu_s, g)
        k3v, k3x = f(v[i-1] + 0.5*k2v, x[i-1] + 0.5*k2x, t[i-1] + 0.5*dt, m, k, mu_s, g)
        k4v, k4x = f(v[i-1] + k3v, x[i-1] + k3x, t[i-1] + dt, m, k, mu_s, g)

        v[i] = v[i-1] + (k1v + 2*k2v + 2*k3v + k4v) / 6
        x[i] = x[i-1] + (k1x + 2*k2x + 2*k3x + k4x) / 6

    return v, x

# Parametri
m = 1.0  # masa tela
k = 0.1  # koeficijent opruge
mu_s = 0.5  # koeficijent statičkog trenja
g = 9.81  # ubrzanje usled gravitacije

# Vremenski korak i vremenski opseg
dt = 0.01
t = np.arange(0, 30, dt)

# Početna brzina i početna deformacija opruge
v0 = 10000.0
x0 = 1.0

# Rešavanje diferencijalne jednačine Runge-Kutta metodom
v, x = runge_kutta_method(differential_equation, v0, x0, t, m, k, mu_s, g)

# Prikaz rezultata
plt.plot(t, v, label='Brzina')
#plt.plot(t, x, label='Deformacija opruge')
plt.xlabel('Vreme (s)')
plt.ylabel('Brzina')
plt.title('Kretanje tela s oprugom i trenjem')
plt.legend()
plt.grid(True)
plt.show()
```
